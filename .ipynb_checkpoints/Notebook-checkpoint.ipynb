### 1. Імпорт данних
[зміст](#0)

```python
url = 'https://archive.ics.uci.edu/static/public/1/abalone.zip'
```

```python
resp = urlopen(url)
myzip = ZipFile(BytesIO(resp.read()))
with myzip.open('abalone.data') as myfile:
    #print(myfile.readline())
    data = pd.read_csv(myfile, names=['Sex',
                                      'Length',
                                      'Diameter',
                                      'Height',
                                      'Whole_weight',
                                      'Shucked_weight',
                                      'Viscera_weight',
                                      'Shell_weight',
                                      'Rings'])
```

### 2. Первинний аналіз данних
[зміст](#0)

```python
data.head()
```

Перевірка дублікатів в датасеті

```python
data.duplicated().sum()
```

```python
data.info()
```

```python
data.describe(include='all')
```

Фіча `Sex` - категоріальна. `Rings` - кількісна, таргет.
Для висоти(`Height`) мінімальне значеня 0?
Подивимось на ці данні.



```python
data[data['Height'] == 0]
```

Для двох особистостей полу `"І"` значеня `Height` дорівнює нулю. Заповнимо їх середнім значеням для цієї статі.

```python
mean_for_I = data[data['Sex'] == 'I']['Height'].mean()
data['Height'] = data['Height'].replace(to_replace=0.0, value=mean_for_I)
```

Подивимось скільи є унікальних таргетів.

```python
data['Rings'].value_counts().count()
```

Візуалізуємо данні, подивимось на росполіл фіч за допомогою `hist()`

```python
data.hist(bins=50, figsize=(12, 12))
plt.show()
```

На графіку `Heigh` бачимо зміщєня вліво через викиди.`Whole_weight`, `Shucked_weight`, `Viscera_weight` та `Shell_weigh` мають асиметричний зсув вліво. `Lenght` та `Diameter` близькі до нормального з невеликим зсувом вправо.

### 3. Фіча інжинірінг
[зміст](#0)<br>

Зробимо копію датасету для додаткового дослідження

```python
data2 = data
```

Перетворимо категоріальну фічу `Sex` в три бінарних `Sex_F`, `Sex_M`, `Sex_I`

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

Подивимось на кореляцію фіч одна-з-одною та з таргетом за допомогою `heatmap()`

```python
sns.set (rc = {'figure.figsize':(8, 8)})
corr_matrix = data.corr()
sns.heatmap(corr_matrix);
```

На графіку бачимо що всі фічі, окрім новоствореної `Sex_*` добре корелюють один з одним. Потрібно позбутися кореляції,

Спробуємо створити новий датасет с декількома новими фічами<br>

| Фіча  | Що означає | Як отримати |
|---    |---         |---
|Volume | Об'єм малюска | Volume = Lenght * Diameter
|Edible_part| Їстівна густина | Edible_part = Whole_weight / Shucked_weight

```python
new_data = pd.DataFrame()
new_data['Volume'] = data['Length'] * data['Diameter']
new_data['Edible_part'] = data['Whole_weight'] / data['Shucked_weight']
new_data['Sex_M'] = data['Sex_M']
new_data['Sex_F'] = data['Sex_F']
new_data['Sex_I'] = data['Sex_I']
new_data['Rings'] = data['Rings']
```

Подивимось на кореляцію новостворених фіч.

```python
sns.set (rc = {'figure.figsize':(8, 8)})
corr_matrix = new_data.corr()
sns.heatmap(corr_matrix);
```

Бачимо що кореляція серед нових фіч меньша за ту що в оригінальному датасеті.

<a name="4"/>

## 4. Масштабування фіч
[зміст](#0)<br>

Сплітимо дані на фічі та таргет

```python
X = new_data.drop('Rings', axis=1).values
y = new_data['Rings'].copy().values
```

Маштабуємо фічі зо допомогою скалєра `MinMaxScaler()`

```python
X_scaled = MinMaxScaler().fit_transform(X)
```

<a name="5"/>

## 5. Поділ датасету на train, valid, test
[зміст](#0)<br>

Поділимо датасет на `train`, `valid` та `test` частини.

```python
X_tr_val, X_test, y_tr_val, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=17)
X_train, X_valid, y_train, y_valid = train_test_split(X_tr_val, y_tr_val, test_size=0.4, random_state=17)
print(f'''All shape: {X_scaled.shape}
X_train:{X_train.shape}, y_train: {y_train.shape}
X_valid:{X_valid.shape}, y_valid: {y_valid.shape}
X_test: {X_test.shape},  y_test:  {y_test.shape}''')
```

<a name="6"/>

### 6. Тренування базової моделі(гіперпараметри=дефолтні)
[зміст](#0)<br>

Побудуємо 3 базові моделі на дефолних параметрах.

```python
# Класифікатор найближчих сусідів
kNN_model_default = KNeighborsClassifier()
kNN_model_default = kNN_model_default.fit(X_train, y_train)
kNN_model_default.score(X_test, y_test)
```

Невеликий `R^2` score. kNN з дефолтними гіперпараметрами не підходить для опису даних.

---

```python
# Класифікатор найбличих сусідів на основі радіусу
rNN_model_default = RadiusNeighborsClassifier()
rNN_model_default = rNN_model_default.fit(X_train, y_train)
rNN_model_default.score(X_test, y_test)
```

Низький `R^2`. Ця модель на дефолтинх гіперпараметрах що гірше описує данні.

---

```python
# Класифікація найближчого центроїда
cNN_model_default = NearestCentroid()
cNN_model_default = cNN_model_default.fit(X_train, y_train)
cNN_model_default.score(X_test, y_test)
```

Взагалі жахливий `R^2`). Сподіваюсь що підбір гіперпараметрів допоможе)

<a name="7"/>

### 7. Подбір гіперпараметрів
[зміст](#0)<br>

По-перше зробимо підбір гіперпараметрів для `KNeighborsClassifier()`

```python
kNN_parameters = {'n_neighbors':[1,3,5,8,10,15,25,50],
                  'weights':    ('uniform','distance'),
                  'algorithm':  ('auto', 'ball_tree', 'kd_tree', 'brute'),
                  'p':          [1,2,3,4]}
kNN = KNeighborsClassifier()
kNN_model_bp = GridSearchCV(kNN, kNN_parameters)
kNN_model_bp = kNN_model_bp.fit(X_valid, y_valid)
```

```python
print(f"Кращій score: {kNN_model_bp.score(X_valid, y_valid)}")
print(f"Кращі гіперпараметри для kNN: {kNN_model_bp.best_params_ }")
```

Після підбору гіперпараметрів результат не видався набагато кращім за модель з дефолтними значенями. Подивимось як справлятся інші моделі, якщо там буде  нормальний score, то все ок.

---

Підбір гіперпараметрів для `RadiusNeighborsClassifier()`

```python
rNN_parameters = {'radius':     [0.1, 0.5, 1, 1.5, 2, 4, 6, 8],
                  'weights':    ('uniform', 'distance'),
                  'algorithm':  ('auto', 'ball_tree', 'kd_tree', 'brute'),
                  'p':          [1,2,3,4]}
rNN = RadiusNeighborsClassifier()
rNN_model_bp = GridSearchCV(rNN, rNN_parameters)
rNN_model_bp = rNN_model_bp.fit(X_train, y_train)
```

```python
print(f"Кращій score: {rNN_model_bp.score(X_valid, y_valid)}")
print(f"Кращі гіперпараметри для rNN: {rNN_model_bp.best_params_}")
```

Тут також не дуже відчутні результати(

---

Підберемо гіперпараметри для `NearestCentroid()`

```python
cNN_parameters = {'metric': ('euclidean', 'braycurtis', 'squareform', 'num_obs_dm('),
                  'shrink_threshold': [0.001, 0.01, 0.1, 1.0, 10.0, 100, ]}
cNN = NearestCentroid()
cNN_model_bp = GridSearchCV(cNN, cNN_parameters)
cNN_model_bp = cNN_model_bp.fit(X_train, y_train)
```

```python
print(f"Кращій score: {cNN_model_bp.score(X_valid, y_valid)}")
print(f"Кращі гіперпараметри для cNN: {cNN_model_bp.best_params_}")
```

<a name="8"/>

### 8. Оцінка результатів
[зміст](#0)<br>

Для оцінки результатів скористаємося функцією метрики `f1_score()`.

```python
def get_report(model, X_test, y_test):
    pred = model.predict(X_test)
    report = f1_score(y_test, pred, average='weighted')

    return report
```

```python
results = pd.DataFrame(columns=('Model', 'F1 score'))
models = [kNN_model_default, rNN_model_default, cNN_model_default, kNN_model_bp, rNN_model_bp, cNN_model_bp]

for model in models:
    rep = get_report(model, X_test, y_test)
    results.loc[len(results.index)] = [model, rep]
```

```python
results
```

У цьому завдані було використано 3 алгоритми класифікації. З дефолтними гіперпараметрами та подібраними. Найшрацім показником виявилась модель `KNeighborsClassifier()` з підібраними гіперпараметрами `{'algorithm': 'auto', 'n_neighbors': 50, 'p': 4, 'weights': 'uniform'}`. Оцінка `F1 score = 0.2361` що э досить низькою. Це може бути пов'язано з дизбалансом класів або з щільніст'ю даних.

### 9. Додаткова класифікація. Target - Sex
[зміст](#0)<br>

```python
data2.head(2)
```

---

Перетворимо таргет `Sex` в кількисний за допомогою `LabelEncoder()`.

```python
label_encoder = LabelEncoder()
data2['Sex'] = label_encoder.fit_transform(data2['Sex'])
data2.head()
```

---

Подивимось на гістограму росподілу `Sex`:

```python
data2['Sex'].hist(figsize=(6, 4));
plt.show()
```

Виходячи з графіку можемо сказати що класів близька кількість і вони сбалансовані.

---

Розділимо данні на фічі та таргет

```python
X_data2 = data2.drop('Sex', axis=1).values
y_data2 = data2['Sex'].copy().values
```

Створимо нові фічі за допомогою `TSNE()`:

```python
tsne = TSNE(n_components=2, verbose=1, random_state=54)
U_data2 = tsne.fit_transform(X_data2)
```

Створимо датасет з новими фічами

```python
new_data2 = pd.DataFrame()
new_data2["y"] = y_data2
new_data2["feature-1"] = U_data2[:,0]
new_data2["feature-2"] = U_data2[:,1]
```

```python
sns.scatterplot(x="feature-1", y="feature-2", hue=new_data2.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=new_data2).set(title="Sex feature in new_data2")
```

На графіку вижно що деякі класи `Sex` знаходятся відокремлено і іх буде легко передбачити за допомогою kNN, а деякі змішані.

---

Розділимо вже новий датасет, з новими згенерованими TSNE фічами на X та y.

```python
X_new = new_data2.drop('y', axis=1).values
y_new = new_data2['y'].copy().values
```

Відмаштабуємо фічі за допомогою `MinMaxScaler()`

```python
X_new_scaled = MinMaxScaler().fit_transform(X_new)
```

Поділимо новий датасет на `train, valid, test`

```python
X_new_tr_val, X_new_test, y_new_tr_val, y_new_test = train_test_split(X_new_scaled, y_new, test_size=0.2, random_state=43)
X_new_train, X_new_valid, y_new_train, y_new_valid = train_test_split(X_new_tr_val, y_new_tr_val, test_size=0.4, random_state=43)
```

Одразу натренеруємо модель з підбором гіперпараметрів.<br>
Для `KNeighborsClassifier()`:

```python
new_kNN_model_bp = GridSearchCV(kNN, kNN_parameters)
new_kNN_model_bp = new_kNN_model_bp.fit(X_new_train, y_new_train)
```

```python
print(f"Кращій score: {new_kNN_model_bp.score(X_new_valid, y_new_valid)}")
print(f"Кращі гіперпараметри для kNN: {new_kNN_model_bp.best_params_ }")
```

Для `RadiusNeighborsClassifier()`:

```python
new_rNN_model_bp = GridSearchCV(rNN, rNN_parameters)
new_rNN_model_bp = new_rNN_model_bp.fit(X_new_train, y_new_train)
```

```python
print(f"Кращій score: {new_rNN_model_bp.score(X_new_valid, y_new_valid)}")
print(f"Кращі гіперпараметри для rNN: {new_rNN_model_bp.best_params_}")
```

Для `NearestCentroid()`:

```python
new_cNN_model_bp = GridSearchCV(cNN, cNN_parameters)
new_cNN_model_bp = new_cNN_model_bp.fit(X_new_train, y_new_train)
```

```python
print(f"Кращій score: {new_cNN_model_bp.score(X_new_valid, y_new_valid)}")
print(f"Кращі гіперпараметри для cNN: {new_cNN_model_bp.best_params_}")
```
